```python
#here we are creating pivot table 
user_movies_df = movies_df.pivot(index='userId',
                                 columns='movie',
                                 values='rating')
```

```python
user_movies_df.head()
```

```python
user_movies_df
```

```python
user_movies_df
```

```python
#Impute those NaNs with 0 values
user_movies_df.fillna(0, inplace=True)
```

```python
user_movies_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances #for calculating distance of similarity
from scipy.spatial.distance import cosine, correlation #for correlation between users 
```

```python
user_sim = 1 - pairwise_distances( user_movies_df.values,metric='cosine') # highly related will be 1 and not highly related will be less than one 
#and we took default one because cos 90 is one 
```

```python
user_sim # we got array of that same dataste 
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = movies_df.userId.unique()
user_sim_df.columns = movies_df.userId.unique()
user_sim_df.columns
```
