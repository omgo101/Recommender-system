```python
#here we are creating pivot table 
user_movies_df = movies_df.pivot(index='userId',
                                 columns='movie',
                                 values='rating')
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances #for calculating distance of similarity
from scipy.spatial.distance import cosine, correlation #for correlation between users 
```

```python
acs=user_movies_df.values
acs
```

```python
acs.mean()
```

```python
import numpy
array = numpy.array(acs)
a=array - 0.7837784856652781
a
```

```python
ser_sim = 1 - pairwise_distances( a,metric='cosine')
ser_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(ser_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = movies_df.userId.unique()
user_sim_df.columns = movies_df.userId.unique()
user_sim_df.columns
```
