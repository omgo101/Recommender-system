```python
#here we are creating pivot table 
user_movies_df = movies_df.pivot(index='userId',
                                 columns='movie',
                                 values='rating')
```

```python
#Impute those NaNs with 0 values
user_movies_df.fillna(0, inplace=True)
```

```python
user_movies_df
```

```python
import numpy as np
from scipy.spatial.distance import jaccard
from sklearn.metrics import jaccard_score
from sklearn.metrics import pairwise_distances #for calculating distance of similarity
from scipy.spatial.distance import cosine, correlation #for correlation between users 
```

```python
jac_sim = 1 - pairwise_distances(user_movies_df.values, metric = "hamming")
```

```python
jac_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(jac_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = movies_df.userId.unique()
user_sim_df.columns = movies_df.userId.unique()
user_sim_df.columns
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(jac_sim, 0) # digonally it will all the  0 because they are same 
user_sim_df.iloc[0:5, 0:5]
```
